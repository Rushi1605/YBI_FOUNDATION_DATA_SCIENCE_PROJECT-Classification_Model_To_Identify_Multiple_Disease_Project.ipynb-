```python
#define y
y = disease['prognosis']
```

```python
#define x
x = disease[['itching','skin_rash','nodal_skin_eruptions','shivering','joint_pain','stomach_pain','acidity','ulcers_on_tongue','blackheads','scurring','skin_peeling','silver_like_dusting','small_dents_in_nails','blister','red_sore_around_nose',]]
```

```python
#split data
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8,random_state=2529)
```

```python
#verify shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
#train model
model.fit(x_train , y_train)
```

```python
#predict with model
y_pred =model.predict(x_test)
```

```python
#model accuracy
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#model evaluation
accuracy_score(y_test , y_pred)
```

```python
#model confusion report
confusion_matrix(y_test, y_pred)
```

```python
#model classification report
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
